```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Cust_Purch_FakeData.csv')
```

```python
data
```

```python
#Total number of data
print('total entries: ', data.size)

#Total number of columns
print('total columns: ', data.shape[1])
```

```python
# max age
print('max age: ', data['age'].max())

#min age
print('min age: ', data['age'].min())

# mean of age
print('mean of age: ', round(data['age'].mean()))
```

```python
#same numbers
duplicate=data['phone'].duplicated(keep=False)
same_number= data[duplicate]
same_number
```

```python
specific_profession = 'Structural Engineer'
engineer = (data['profession'] == specific_profession).sum()
print('total number: ', engineer)
```

```python
#strectural male engineer

male_engineer = ((data['profession']=='Structural Engineer') & (data['gender'] == 'Male')).sum()
print('total male strectural engineer: ', male_engineer)
```

```python
female_engineer = ((data['profession']=='Structural Engineer') & (data['gender'] == 'Male') & (data ['province'] == 'AB')).sum()
print('total female strectural engineer: ', female_engineer)
```

```python
# max age
print('max age: ', data['price(CAD)'].max())

#min age
print('min age: ', data['price(CAD)'].min())

# mean of age
print('mean of age: ', round(data['price(CAD)'].mean()))
```

```python
#encourage the customer

no_spending = ((data['price(CAD)'] == 0) | (data['price(CAD)'].isnull())).sum()
no_spending
```

```python
#thanks coupon
reward = (data['price(CAD)'] >= 100).sum()
print('Thank You: ', reward)
```

```python
#associated with credit card 
card_number = '5020000000000230'
emails_with_card=data[data['cc_no'] == card_number] ['email']
email_count=len(emails_with_card)
email_count
print('email associate with cards', emails_with_card.tolist()),
print(f"Total emails associate with card: {email_count}")
```

```python
#visa user
visa_user =data[data['cc_type']== 'Visa']
total_visa_user = len(visa_user)
print(f"Number of people using Visa Card: {total_visa_user}")
```

```python
spending =data[(data['cc_type']== 'Visa') & (data ['price(CAD)'] >= 100)]
total_cad = len(spending)
print(f"Number of people using Visa Card: {total_cad}")
```

```python
data['email_provider']= data['email'].apply(lambda x: x.split('@')[-1])
top_provider=data['email_provider'].value_counts().head(5)
top_provider
```
